```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons, make_blobs
from sklearn.datasets import make_classification
from sklearn.datasets import make_biclusters

# Load the datasets
datasets = {
    'noisy_circles': make_circles(n_samples=1000, factor=0.5, noise=0.1),
    'noisy_moons': make_moons(n_samples=1000, noise=0.1),
    'blobs': make_blobs(n_samples=1000, centers=3, random_state=42),
    'aniso': make_blobs(n_samples=1000, centers=3, random_state=42),
    'no_structure': make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1),
}

# Define colors for each dataset
colors = {
    'noisy_circles': 'blue',
    'noisy_moons': 'orange',
    'blobs': 'green',
    'aniso': 'red',
    'no_structure': 'purple',
}
```

```python
# Plot each dataset
plt.figure(figsize=(15, 10))
for i, (name, data) in enumerate(datasets.items(), 1):
    X, y = data
    plt.subplot(2, 3, i)
    plt.scatter(X[:, 0], X[:, 1], c=colors[name], edgecolors='k', alpha=0.6)
    plt.title(name)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets using pandas
datasets = {
    'noisy_circles': pd.read_csv("noisy_circles.txt", header=None, names=['x', 'y']),
    'noisy_moons': pd.read_csv("noisy_moons.txt", header=None, names=['x', 'y']),
    'blobs': pd.read_csv("blobs.txt", header=None, names=['x', 'y']),
    'aniso': pd.read_csv("aniso.txt", header=None, names=['x', 'y']),
    'no_structure': pd.read_csv("no_structure.txt", header=None, names=['x', 'y']),
}

# Define colors for each dataset
colors = {
    'noisy_circles': 'blue',
    'noisy_moons': 'orange',
    'blobs': 'green',
    'aniso': 'red',
    'no_structure': 'purple',
}

# Plot each dataset
plt.figure(figsize=(15, 10))
for i, (name, data) in enumerate(datasets.items(), 1):
    plt.subplot(2, 3, i)
    plt.scatter(data['x'], data['y'], c=colors[name], edgecolors='k', alpha=0.6)
    plt.title(name)
plt.tight_layout()
plt.show()
```
