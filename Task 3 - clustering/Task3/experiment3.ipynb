# Machine Learning Task 3 Clustering

## 1. Import necessary library and Load 5 datasets 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_circles, make_moons, make_blobs
from sklearn.datasets import make_classification
from sklearn.datasets import make_biclusters
from sklearn.cluster import KMeans, DBSCAN

# 读取数据集
noisy_circles = pd.read_table('noisy_circles.txt', header=None, sep =' ')
noisy_moons = pd.read_table('noisy_moons.txt', header=None, sep =' ')
blobs = pd.read_table('blobs.txt', header=None, sep =' ')
aniso = pd.read_table('aniso.txt', header=None, sep =' ')
no_structure = pd.read_table('no_structure.txt', header=None, sep =' ')

# 定义每个数据集的颜色
colors = {
    'noisy_circles': 'blue',
    'noisy_moons': 'orange',
    'blobs': 'green',
    'aniso': 'red',
    'no_structure': 'purple',
}
```

## 2. Drawing scatterplots to visualize data sets 

```python
# 绘制每个数据集
plt.figure(figsize=(15, 10))

# 读取并绘制 Noisy Circles 数据集
plt.subplot(2, 3, 1)
plt.scatter(noisy_circles.iloc[:, 0], noisy_circles.iloc[:, 1], c=colors['noisy_circles'], s=10)
plt.title("Noisy Circles (n_clusters=2)")

# 读取并绘制 Noisy Moons 数据集
plt.subplot(2, 3, 2)
plt.scatter(noisy_moons.iloc[:, 0], noisy_moons.iloc[:, 1], c=colors['noisy_moons'], s=10)
plt.title("Noisy Moons (n_clusters=2)")

# 读取并绘制 Blobs 数据集
plt.subplot(2, 3, 3)
plt.scatter(blobs.iloc[:, 0], blobs.iloc[:, 1], c=colors['blobs'], s=10)
plt.title("Blobs (n_clusters=3)")

# 读取并绘制 Anisotropicly Distributed Data 数据集
plt.subplot(2, 3, 4)
plt.scatter(aniso.iloc[:, 0], aniso.iloc[:, 1], c=colors['aniso'], s=10)
plt.title("Anisotropicly Distributed Data (n_clusters=3)")

# 读取并绘制 No Structure 数据集
plt.subplot(2, 3, 5)
plt.scatter(no_structure.iloc[:, 0], no_structure.iloc[:, 1], c=colors['no_structure'], s=10)
plt.title("No Structure (n_clusters=3)")

plt.tight_layout()
plt.show()
```

## 3. Use K-means and DBScan to cluster 5 datasets.

### (1). Cluster on noisy_circles.txt

```python
# K-means 聚类
noisy_circles_kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
y_pred_kmeans1 = noisy_circles_kmeans.fit_predict(noisy_circles)

# 绘制 K-means 聚类结果
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(noisy_circles.iloc[:, 0], noisy_circles.iloc[:, 1], c=y_pred_kmeans1, s=10, cmap='viridis')
plt.title("Noisy Circles - K-means (n_clusters=2)")

# DBSCAN 聚类
noisy_circles_dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred_dbscan1 = noisy_circles_dbscan.fit_predict(noisy_circles)

# 绘制 DBSCAN 聚类结果
plt.subplot(1, 2, 2)
plt.scatter(noisy_circles.iloc[:, 0], noisy_circles.iloc[:, 1], c=y_pred_dbscan1, s=10, cmap='viridis')
plt.title("Noisy Circles - DBSCAN")

plt.tight_layout()
plt.show()
```

### (2). Cluster on noisy_moons.txt

```python
# K-means 聚类
noisy_moons_kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
y_pred_kmeans2 = noisy_moons_kmeans.fit_predict(noisy_moons)

# 绘制 K-means 聚类结果
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(noisy_moons.iloc[:, 0], noisy_moons.iloc[:, 1], c=y_pred_kmeans2, s=10, cmap='viridis')
plt.title("Noisy Moons - K-means (n_clusters=2)")

# DBSCAN 聚类
noisy_moons_dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred_dbscan2 = noisy_moons_dbscan.fit_predict(noisy_moons)

# 绘制 DBSCAN 聚类结果
plt.subplot(1, 2, 2)
plt.scatter(noisy_moons.iloc[:, 0], noisy_moons.iloc[:, 1], c=y_pred_dbscan2, s=10, cmap='viridis')
plt.title("Noisy Moons - DBSCAN")

plt.tight_layout()
plt.show()
```

### (3). Cluster on blobs.txt

```python
# K-means 聚类
blobs_kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_pred_kmeans3 = blobs_kmeans.fit_predict(blobs)

# 绘制 K-means 聚类结果
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(blobs.iloc[:, 0], blobs.iloc[:, 1], c=y_pred_kmeans3, s=10, cmap='viridis')
plt.title("Blobs - K-means (n_clusters=3)")

# DBSCAN 聚类
blobs_dbscan = DBSCAN(eps=1.5, min_samples=5)
y_pred_dbscan3 = blobs_dbscan.fit_predict(blobs)

# 绘制 DBSCAN 聚类结果
plt.subplot(1, 2, 2)
plt.scatter(blobs.iloc[:, 0], blobs.iloc[:, 1], c=y_pred_dbscan3, s=10, cmap='viridis')
plt.title("Blobs - DBSCAN")

plt.tight_layout()
plt.show()
```

### (4). Cluster on aniso.txt

```python
# K-means 聚类
aniso_kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_pred_kmeans4 = aniso_kmeans.fit_predict(aniso)

# 绘制 K-means 聚类结果
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(aniso.iloc[:, 0], aniso.iloc[:, 1], c=y_pred_kmeans4, s=10, cmap='viridis')
plt.title("Anisotropicly Distributed Data - K-means (n_clusters=3)")

# DBSCAN 聚类
aniso_dbscan = DBSCAN(eps=0.5, min_samples=5)
y_pred_dbscan4 = aniso_dbscan.fit_predict(aniso)

# 绘制 DBSCAN 聚类结果
plt.subplot(1, 2, 2)
plt.scatter(aniso.iloc[:, 0], aniso.iloc[:, 1], c=y_pred_dbscan4, s=10, cmap='viridis')
plt.title("Anisotropicly Distributed Data - DBSCAN")

plt.tight_layout()
plt.show()
```

```python
# K-means 聚类
no_structure_kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_pred_kmeans5 = no_structure_kmeans.fit_predict(no_structure)

# 绘制 K-means 聚类结果
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(no_structure.iloc[:, 0], no_structure.iloc[:, 1], c=y_pred_kmeans5, s=10, cmap='viridis')
plt.title("No Structure - K-means (n_clusters=3)")

# DBSCAN 聚类
no_structure_dbscan = DBSCAN(eps=0.5, min_samples=5)
y_pred_dbscan5 = no_structure_dbscan.fit_predict(no_structure)

# 绘制 DBSCAN 聚类结果
plt.subplot(1, 2, 2)
plt.scatter(no_structure.iloc[:, 0], no_structure.iloc[:, 1], c=y_pred_dbscan5, s=10, cmap='viridis')
plt.title("No Structure - DBSCAN")

plt.tight_layout()
plt.show()
```
